```python
#Splitting the Dataset into Test and Train
import os
from glob import glob
import shutil
from sklearn.model_selection import train_test_split
# do test train splitting

# find image names
image_files = glob("/content/BloodRBC-Detection-Dataset-Processed/*.jpeg")
# remove file extension
image_names = [name.replace(".jpeg","") for name in image_files]
# Use scikit learn function for convenience
train_names, test_names = train_test_split(image_names, test_size=0.2)
```

```python
test_dir = "/content/gdrive/MyDrive/Projects/blood_cell/updated/BloodRBC_Dataset/Testing"
train_dir = "/content/gdrive/MyDrive/Projects/blood_cell/updated/BloodRBC_Dataset/Training"
```

```python
def batch_move_files(file_list, destination_path):
    for file in file_list:
         image = file + '.jpeg'
         xml = file + '.xml'
         shutil.copy(image, destination_path)
         shutil.copy(xml, destination_path)
    return


batch_move_files(test_names, test_dir)
batch_move_files(train_names, train_dir)
```
